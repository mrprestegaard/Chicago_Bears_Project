```python
# Import our data processing library (note: you may have to install this!)
import pandas as pd
import altair as alt
```

```python
# Let's use this to upload a sample dataset and show the start of the dataset
filepath = "/Users/madelineprestegaard/Desktop/sportsref_Bears.csv"
data= pd.read_csv(filepath)
data.head()
```

```python
# Assuming `data` is your DataFrame
filtered_data = data[(data['Year'] >= 2000) & (data['Year'] <= 2025)]

# Define the years in which the division was won
division_won_years = [2001, 2005, 2006, 2010, 2018]

# Add a new column for color encoding and a custom message for division-winning years
filtered_data['ChampionshipWon'] = filtered_data['Year'].isin(division_won_years)
filtered_data['ChampionshipWonText'] = filtered_data['ChampionshipWon'].apply(lambda x: 'Championship Won' if x else '')

# Create the chart with conditional coloring
chart = alt.Chart(filtered_data).mark_bar().encode(
    x=alt.X('Year:Q', scale=alt.Scale(domain=[2000, 2025]), title="Year"),
    y=alt.Y('W:Q', title="Wins"),
    color=alt.condition(
        alt.datum.DivisionWon,  # Condition based on the new column
        alt.value('green'),  # Color for bars that meet the condition
        alt.value('lightgreen') # Color for other bars
    ),
    tooltip=["Year", "W", "L", alt.Tooltip('ChampionshipWonText:N', title="Status")]
).properties(
    title="Wins by Year (2000-2025)"
)

chart
```
